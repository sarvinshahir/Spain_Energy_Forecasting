```python
import pandas as pd

energy = pd.read_csv('energy_dataset.csv')
weather = pd.read_csv('weather_features.csv')

print("Energy shape:", energy.shape)
print("Weather shape:", weather.shape)

energy.info()
weather.info()
```

```python

# --- Convert timestamps ---
energy['time'] = pd.to_datetime(energy['time'], errors='coerce')
weather['dt_iso'] = pd.to_datetime(weather['dt_iso'], errors='coerce')

# --- Drop fully empty columns ---
energy = energy.drop(columns=['generation hydro pumped storage aggregated',
                              'forecast wind offshore eday ahead'])

# --- Fill missing values using the new syntax ---
energy = energy.ffill()

# --- Ensure 'time' is datetime (no tz info) ---
if pd.api.types.is_datetime64_any_dtype(energy['time']):
    energy['time'] = energy['time'].dt.tz_localize(None)
else:
    print("âš ï¸ 'time' column is not datetime; check conversion:")
    print(energy['time'].head())

# --- Same for weather ---
if pd.api.types.is_datetime64_any_dtype(weather['dt_iso']):
    weather['dt_iso'] = weather['dt_iso'].dt.tz_localize(None)
```

```python
print(energy['time'].dtype)
print(weather['dt_iso'].dtype)
```

```python
import pandas as pd

def ensure_datetime(series):
    # Convert everything to string first, then parse; coerce errors to NaT
    s = series.astype(str)
    s = pd.to_datetime(s, utc=True, errors='coerce')
    return s.dt.tz_convert(None)

energy['time'] = ensure_datetime(energy['time'])
weather['dt_iso'] = ensure_datetime(weather['dt_iso'])

print(energy['time'].dtype)
print(weather['dt_iso'].dtype)
```

# Aggregated data

```python
# Aggregate weather data (averaged across cities)
weather_agg = (
    weather.groupby('dt_iso')
    .agg({
        'temp': 'mean',
        'humidity': 'mean',
        'pressure': 'mean',
        'wind_speed': 'mean',
        'rain_1h': 'sum',
        'clouds_all': 'mean'
    })
    .reset_index()
    .rename(columns={'dt_iso': 'time'})
)

# Merge energy and aggregated weather
df = pd.merge(energy, weather_agg, on='time', how='inner')
print("Merged shape:", df.shape)

# Quick check
df.info()
df.head()
```

```python
df.describe([0.01, 0.5, 0.99]).T
```

```python
zero_columns = [
    'generation fossil coal-derived gas',
    'generation fossil oil shale',
    'generation fossil peat',
    'generation geothermal',
    'generation marine',
    'generation wind offshore'
]

df.drop(columns=zero_columns, inplace=True)
```

```python
df.head()
```

```python

# Mark outliers above 1100 hPa as NaN
df.loc[df['pressure'] > 1100, 'pressure'] = np.nan

# Interpolate or forward-fill them
df['pressure'] = df['pressure'].interpolate()

# Check the new range
print(df['pressure'].describe())
```

# 6. Correlation Between Load, Weather, and Prices

```python
corr_cols = [
    'total load actual', 'price actual',
    'generation solar', 'generation wind onshore',
    'generation fossil gas', 'temp', 'humidity',
    'wind_speed', 'rain_1h'
]
corr = df[corr_cols].corr()

plt.figure(figsize=(9,7))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.tight_layout()
plt.show()
```

```python
numeric_df = df.select_dtypes(include=['float64', 'int64'])

# Compute correlation
corr = numeric_df.corr()

# Plot
plt.figure(figsize=(16, 14))
sns.heatmap(
    corr,
    cmap='coolwarm',
    center=0,
    cbar_kws={'label': 'Correlation Coefficient'},
    square=True,
    xticklabels=True,
    yticklabels=True
)
plt.title('Correlation Heatmap: All Numeric Features', fontsize=14)
plt.tight_layout()
plt.show()
```
